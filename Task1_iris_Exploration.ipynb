## 1. Problem Statement

The objective of this task is to explore and visualize the Iris dataset to understand
data distributions, relationships between features, and potential patterns among different species.

Through exploratory data analysis (EDA), we aim to:
- Understand dataset structure
- Analyze statistical summaries
- Visualize feature relationships
- Detect possible outliers


```python
# Suppress warnings for clean output
import warnings
warnings.filterwarnings("ignore")

# Data manipulation
import numpy as np
import pandas as pd

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Plot styling
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
# Load Iris dataset from seaborn
df = sns.load_dataset("iris")

# Display first 5 rows
df.head()
```

```python
df.shape
```

```python
df.info()

```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
df.hist(bins=15, figsize=(10,8))
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Features")
plt.show()
```

## Final Conclusion

The Iris dataset contains 150 observations with 4 numerical features and 1 categorical target variable (species).

Key findings:
- The dataset has no missing values.
- Petal length and petal width show strong separation among species.
- Sepal width shows higher variability compared to other features.
- No extreme outliers significantly affecting the dataset.

Overall, the dataset is clean, balanced, and well-suited for further machine learning tasks.
